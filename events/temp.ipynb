```python
import json
import os
import pandas as pd

def load_jsonl_data(path):
    data = []
    with open(path) as reader:
        for row in reader:
            d = json.loads(row)
            data.append(d)
    return data
```

```python
data = load_jsonl_data('/inspire/hdd/project/socialsimulation/linjiayu-CZXS25120090/FYDUAN/data/events_labeled_data_1/bucket_id=100/events_raw.jsonl')

posts = []
n = 0
for x in data:
    x['posts'] = [xx for xx in x['posts'] if xx['event']]
    
    if len(x['posts']) > 10:
        print(len(x['posts']))
        posts.extend(x['posts'])

        n += 1

print(n)
df = pd.DataFrame(posts)
df['is_offline_events'] = ''
df['precise_label'] = ''

df.to_excel('/inspire/hdd/project/socialsimulation/linjiayu-CZXS25120090/FYDUAN/exp_results/event_label/events.xlsx')
```
