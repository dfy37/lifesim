```python
import json
import os
import pandas as pd

def load_jsonl_data(path):
    data = []
    with open(path) as reader:
        for row in reader:
            d = json.loads(row)
            data.append(d)
    return data
```

```python
data = load_jsonl_data('/inspire/hdd/project/socialsimulation/linjiayu-CZXS25120090/FYDUAN/data/events_labeled_data/bucket_id=2/events_raw.jsonl')

df = pd.DataFrame(data)
df['is_offline_events'] = ''
df['precise_label'] = ''

df.to_excel('/inspire/hdd/project/socialsimulation/linjiayu-CZXS25120090/FYDUAN/exp_results/event_label/events.xlsx')
```
